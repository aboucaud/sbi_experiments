Here we show you how to access the various distributions we are using for developing our models

```python
from sbiexpt import distributions
```

#### Two moons

```python
dist = distributions.get_two_moons(sigma=0.1) # Instantiate distribution

# Create a visualisation grid
coords = jnp.stack(jnp.meshgrid(jnp.linspace(-1.5,2.5,128),
                                jnp.linspace(-1,1.5,128)),
                   axis=-1)
# Plot density
contourf(coords[...,0], coords[...,1], dist.prob(coords), 100);
```

#### Swiss roll

```python
dist = distributions.get_swiss_roll(sigma=1) # Instantiate distribution

# Create a visualisation grid
coords = jnp.stack(jnp.meshgrid(jnp.linspace(-15,15,128),
                                jnp.linspace(-15,15,128)),
                   axis=-1)
# Plot density
contourf(coords[...,0], coords[...,1], dist.prob(coords), 100);
```
